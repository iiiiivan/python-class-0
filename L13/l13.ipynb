```python
def make_bar_graph(labels, bar_heights, title):
    plt.style.use('fivethirtyeight')
    x_position=range(len(bar_heights))
    plt.title(title)
    plt.xticks(x_position, labels)
    figure=plt.bar(x_position, bar_heights)
```

```python
make_bar_graph(["monday", "tuesday"], [5, 10], "students")
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-deep')

# make data
np.random.seed(3)
x = 0.5 + np.arange(8)
y = np.random.uniform(2, 7, len(x))

# plot
fig, ax = plt.subplots()

ax.stem(x, y)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))
plt.xticks(x, list(range(len(x))))
plt.title('stem graph')
plt.show()
```

```python
def make_stem_graph(labels, stem_height, title, draw_labels=True):
    plt.style.use('seaborn-deep')

    # make data
    
    x = range(len(stem_height))

    # plot
    fig, ax = plt.subplots()

    ax.stem(x, stem_height)

    if draw_labels:
        plt.xticks(x, labels)
    else:
        plt.xticks([])
    plt.title(title)
    plt.show()
```

```python
make_stem_graph([2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021], [1,2,3,4,5,6,7,8,9], 'title')
```

```python
make_stem_graph([2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021], [1,2,3,4,5,6,7,8,9], 'title', draw_labels=False)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from random import  random
from math import pi, sin, cos

x=[]
y=[]
for _ in range(1000):
    radians=random()*pi
    radius=random()*2 - 1
    x.append(radius*cos(radians))
    y.append(radius*sin(radians))

# plot:
fig, ax = plt.subplots()

ax.hexbin(x, y, gridsize=20)

ax.set(xlim=(-1, 1), ylim=(-1, 1))
plt.xticks([])
plt.yticks([])
plt.show()
```
